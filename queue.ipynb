#### Очередь

```python
class Queue:
    def __init__(self):
        self.items = []
    
    def isEmpty(self):
        return self.items == []
    
    def enqueue(self, item):
        self.items.insert(0, item)

    def dequeue(self):
        return self.items.pop()
    
    def size(self):
        return len(self.items)
```

```python
class PriorityQueue:
    def __init__(self):
        self.items = []
    
    def isEmpty(self):
        return self.items == []
    
    def enqueue(self, value, priority):
        if self.items == []:
            self.items.append([value, priority])
        else:
            for i in range(len(self.items)):
                if priority == self.items[i][1]:
                    self.items.insert(i, [value, priority])
                    break
                elif priority > self.items[i][1]:
                    try:
                        if priority > self.items[i + 1][1]:
                            continue
                        elif priority == self.items[i + 1][1]:
                            self.items.insert(i + 1, [value, priority])
                        else:
                            self.items.append([value, priority])
                    except:
                        self.items.append([value, priority])
                    break
                
    def dequeue(self):
        self.items.pop()
    
    def dequeue2(self):
        self.items = self.items[:-1]
    
    def dequeue3(self):
        del self.items[-1]
    
    def show(self):
        return self.items
    
    def clear(self):
        self.items = []
        print('error')

    def size(self):
        return len(self.items)
```

```python
q = PriorityQueue()
q.enqueue(1, 1)
print(q.items)
q.enqueue(2, 1)
print(q.items)
q.enqueue(3, 2)
print(q.items)
q.enqueue(2, 2)
print(q.items)
q.enqueue(2, 1)
print(q.items)
q.enqueue(3, 2)
print(q.items)
q.enqueue(2, 2)
print(q.items)
q.enqueue(5, 5)
print(q.items)
```

#### Задача на массив

```python
data = [
    1,
    5,
    8,
    3,
    11,
    1,
    4,
    9,
    2,
    10,
    4
]
def maxmult(data):
    maxmuls = [[data[0] * data[1], data[0], data[1]]]
    for i in range(1, len(data)-1):
        if data[i] * data[i + 1] > maxmuls[0][0]:
            maxmuls = [[data[i] * data[i + 1], data[i], data[i + 1]]]
        elif data[i] * data[i + 1] == maxmuls[0][0]:
            maxmuls.append([data[i] * data[i + 1], data[i], data[i + 1]])

    maxmuls_sort1 = sorted(maxmuls, key = lambda x: x[1])
    maxmuls_sort2 = sorted(maxmuls, key = lambda x: x[2])

    if maxmuls_sort1[0][1] < maxmuls_sort2[0][2]:
        return maxmuls_sort1[0][1], maxmuls_sort1[0][2] 
    else:
        return maxmuls_sort2[0][1], maxmuls_sort2[0][2]
```

```python
maxmult(data)
```

```python
assert maxmult(data) == (10, 4)
assert maxmult([1, 1, 1]) == (1, 1)
assert maxmult([3, 3, 2]) == (3, 3)
assert maxmult([4, 1, 3, 2]) == (3, 2)
```
